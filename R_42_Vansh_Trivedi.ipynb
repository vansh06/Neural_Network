```python
#Mp 
from sklearn.metrics import accuracy_score
import numpy as np
class MP_Neuron:
  def __intit__(self):
    self.b=0
  def model(self,x):
    return np.sum(x)>=self.b
  def fit(self,x,y):
    accuracy={}
    for b in range(x.shape[1]+1):
      self.b=b
      yh=[]
      for row in x:
        yh.append(self.model(row))
      print(b,len(yh)),len(y)
      accuracy[b]=accuracy_score(yh,y)
    best_b=max(accuracy,key=accuracy.get)
    self.b=best_b
    return [accuracy,best_b,accuracy[best_b]]
  def predict(self,x,y):
    yh=[]
    for row in x:
      yh.append(self.model(row))
    accuracy=accuracy_score(y,yh)
    return accuracy
```

```python
import numpy as np

inputs = np.array([[0,0],[1,0],[0,1],[1,1]])
Y = np.array([[0],[1],[1],[0]])

epochs=1000
eta=0.9
input_N, hidden_N, output_N = 2,2,1
W_H = [[0.5,-0.5],[-0.5,1.0]]
B_H =[[0.5,0.2]]
W_Out = np.random.uniform(size=(hidden_N,output_N))
B_Out = np.random.uniform(size=(1,output_N))

def sigmoid (x):
    return 1/(1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)

for i in range(epochs):
  hidden_layer_activation = np.dot(inputs,W_H)
  hidden_layer_activation += B_H
  hidden_layer_output = sigmoid(hidden_layer_activation)
  output_layer_activation = np.dot(hidden_layer_output,W_Out)
  output_layer_activation += B_Out
  Y_pred = sigmoid(output_layer_activation)


  #Backpropagation
  error = Y - Y_pred
  d_pred_Y = error * sigmoid_derivative(Y_pred)
  error_hidden_layer = d_pred_Y.dot(W_Out.T)
  d_hidden_layer = error_hidden_layer * sigmoid_derivative(hidden_layer_output)
 
 
  #Updating Weights and Biases
  W_Out+= eta*hidden_layer_output.T.dot(d_pred_Y)
  B_Out += eta*np.sum(d_pred_Y,axis=0,keepdims=True)
  W_H += eta*inputs.T.dot(d_hidden_layer)
  B_H+= eta*np.sum(d_hidden_layer,axis=0,keepdims=True)

print("Final hidden weights: ",end=' ')
print(*W_H)
print("\nFinal hidden bias: ",end=' ')
print(*B_H)
print("\nFinal output weights: ",end=' ')
print(*W_Out)
print("\nFinal output bias: ",end=' ')
print(*B_Out)
print("\nPredicted output after 1000 epochs: ",end=' ')
print(*Y_pred)
```
